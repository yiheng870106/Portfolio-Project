The investment universe includes COST, GLD, TSM, VNQ, VXUS, XLE, XLU, and XLV.

1. **COST**: Costco Wholesale Corporation, a large U.S. retailer .

2. **GLD**: SPDR Gold Shares, an ETF designed to provide exposure to the price of gold.

3. **TSM**: Taiwan Semiconductor Manufacturing Company, the world’s leading pure-play semiconductor foundry and a key company in the global chip supply chain.

4. **VNQ**: Vanguard Real Estate ETF, an ETF that primarily invests in U.S. REITs and real-estate-related companies.

5. **VXUS**: Vanguard Total International Stock ETF, an ETF that tracks non-U.S. developed and emerging market equities.

6. **XLE**: Energy Select Sector SPDR Fund, an ETF including oil, gas, and energy equipment companies.

7. **XLU**: Utilities Select Sector SPDR Fund, an ETF representing the utilities.

8. **XLV**: Health Care Select Sector SPDR Fund, an ETF including pharmaceuticals, medical devices, health care providers, and biotechnology firms.

Download the daily prices from https://bulkstockdatadownloader.app. The chart below shows the daily prices of the nine selected stocks/ETFs from 01/01/2025 to 12/31/2025

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel("3F6G3GAX.xlsx")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print("Daily Prices from 01/01/2025 to 12/31/2025:\n",df)

plt.figure(figsize=(15,7))
for col in df.columns:
    plt.plot(df.index, df[col], label=col)
plt.legend()
plt.title("Daily Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

Compute both arithmetic and log returns over the one year period.

```python
arith_returns = df.pct_change().dropna()
# log_returns = np.log(df/df.shift(1))

print("Arithmetic Returns (daily):\n")
print(arith_returns)
# print("\nLog Returns:\n")
# print(log_returns)

# plt.figure(figsize=(12,8))
# for col in df.columns:
#     plt.plot(arith_returns.index, arith_returns[col], label=f"{col} - Arithmetic", alpha=0.6)

# plt.title("Arithmetic Returns (Daily)")
# plt.xlabel("Date")
# plt.ylabel("Daily Return")
# plt.legend(loc="upper right", fontsize=8, ncol=2)
# plt.grid(True)
# plt.show()
```

Compute the expected returns for individual stocks/ETFs and the covariance matrix for returns.

```python
exp_arith_returns = arith_returns.mean()
# exp_log_returns = log_returns.mean()
var = arith_returns.var()
cov_matrix = arith_returns.cov()

print("\nExpected Arithmetic Returns (daily):\n")
print(exp_arith_returns)
# print("\nExpected Log Returns:\n")
# print(exp_log_returns)
print("\nVariance of Arithmetic Returns (daily):\n")
print(var)
print("\nCovariance Matrix of Arithmetic Returns:\n")
print(cov_matrix)
```

Find:

(a). The global minimum variance portfolio;

(b). The market portfolio based on these nine stocks/ETFs, assuming the annual risk-free rate r = 4%

```python
C=cov_matrix
C_inv = np.linalg.inv(C)
u = np.ones(len(C))
m = exp_arith_returns

w_MVP = C_inv @ u / (u.T @ C_inv @ u)
print("Global minimum variance portfolio (MVP):\n [COST, GLD, TSM, VNQ, VXUS, XLE, XLU, XLV] = \n", w_MVP)
print("\n")

rf_annual = 0.04
rf_daily = (1 + rf_annual)**(1/252) - 1

print("daily risk-free return", rf_daily)
print("\n")

w_M = C_inv @ (m-rf_daily*u) / (u.T @ C_inv @ (m-rf_daily*u))
print("The market portfolio:\n [COST, GLD, TSM, VNQ, VXUS, XLE, XLU, XLV] = \n", w_M)
```

Backtest the portfolio values over the past three month (10/2026 - 12/2026).

```python
end_date = df.index.max()
start_date = end_date - pd.DateOffset(months=3)

df_last3m = df.loc[start_date:end_date]
exp_return_last3m = df_last3m.pct_change().mean()
var_last3m = df_last3m.pct_change().var()

print("Over the past 3 months")
for col in df_last3m.columns:
  print(f"Expected return of {col}: {exp_return_last3m[col]:.6f}")
  print(f"Variance of {col}: {var_last3m[col]:.6f}\n")
```

```python
ret_MVP = arith_returns @ w_MVP
ret_M = arith_returns @ w_M

ret_MVP_last3m = ret_MVP.loc[start_date:end_date]
ret_M_last3m = ret_M.loc[start_date:end_date]

val_MVP = (1 + ret_MVP).cumprod()
val_M = (1 + ret_M).cumprod()

MVP_last3m = val_MVP.loc[start_date:end_date]
M_last3m = val_M.loc[start_date:end_date]

plt.figure(figsize=(10,5))
plt.plot(MVP_last3m.index, MVP_last3m.values, label='MVP Portfolio')
plt.plot(M_last3m.index, M_last3m.values, label='Market Portfolio')
plt.title("Daily Prices (last 3 months)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```

Compute the portfolio return and the portfolio volatility.

```python
exp_MVP = ret_MVP_last3m.mean()
var_MVP = ret_MVP_last3m.var()
exp_M = ret_M_last3m.mean()
var_M = ret_M_last3m.var()

print(f"Expected return of MVP portfolio: {exp_MVP:.6f}")
print(f"Variance of MVP portfolio: {var_MVP:.6f}")
print(f"\nExpected return of market portfolio: {exp_M:.6f}")
print(f"Variance of market portfolio: {var_M:.6f}")
```

Assuming a risk-free rate of 4%, the Sharpe ratios are calculated and summarized in the table below.

```python
sharpe_ratio = []
for i in range(len(x)):
  sharpe_ratio.append((y[i]-rf_daily)/x[i])

df_sharpe = pd.DataFrame({
    'Asset': labels,
    'Expected Return': y,
    'Variance': var,
    'Std Dev': x,
    'Sharpe Ratio': sharpe_ratio
})
print("Assume the annual risk-free rate is 4%.")
df_sharpe = df_sharpe.round(6)
print(df_sharpe)
```
